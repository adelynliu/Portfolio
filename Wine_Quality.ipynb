```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
wine_quality = fetch_ucirepo(id=186)

# data (as pandas dataframes)
X = wine_quality.data.features
y = wine_quality.data.targets

# Combine X and y into one DataFrame
df = pd.concat([X, pd.DataFrame(y, columns=['quality'])], axis=1)
```

```python
#check the columns that have potential outliners
columns_to_check = ['residual_sugar', 'free_sulfur_dioxide', 'total_sulfur_dioxide' ]

# Calculate z-scores for 'residual_sugar', 'free_sulfur_dioxide', 'total_sulfur_dioxide'
z_scores_sugar = stats.zscore(df['residual_sugar'])
z_scores_free = stats.zscore(df['free_sulfur_dioxide'])
z_scores_total = stats.zscore(df['total_sulfur_dioxide'])

# Define a threshold for considering data as outliers (e.g., 3 standard deviations)
threshold = 3

# Remove outliers
df_new = df[(np.abs(z_scores_sugar) < threshold)]
df_new = df_new[(np.abs(z_scores_free) < threshold)]
df_new= df_new[(np.abs(z_scores_total) < threshold)]

df_new.describe().T
```

## Fit an Explainable Boosting Machine using interpret

```python
# Splitting the data into training, validation, and test sets
train_data, test_data = train_test_split(df_new, test_size=0.2, random_state=42)
train_data, val_data = train_test_split(train_data, test_size=0.2, random_state=42)

# Selecting features and target variable for training set
X_train = train_data[['alcohol', 'volatile_acidity', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'residual_sugar', 'density', 'sulphates', 'chlorides', 'pH', 'fixed_acidity']]
y_train = train_data['quality']

# Selecting features and target variable for validation set
X_val = val_data[['alcohol', 'volatile_acidity', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'residual_sugar', 'density', 'sulphates', 'chlorides', 'pH', 'fixed_acidity']]
y_val = val_data['quality']

# Selecting features and target variable for test set
X_test = test_data[['alcohol', 'volatile_acidity', 'free_sulfur_dioxide', 'total_sulfur_dioxide', 'residual_sugar', 'density', 'sulphates', 'chlorides', 'pH', 'fixed_acidity']]
y_test = test_data['quality']


ebm = ExplainableBoostingRegressor()
ebm.fit(X_train, y_train)

# Predicting on the validation set
y_val_pred = ebm.predict(X_val)

# Calculating the mean squared error on the validation set
mse_val = mean_squared_error(y_val, y_val_pred)
print(f"Mean Squared Error on Validation Set: {mse_val}")

# Make predictions on the test set
y_test_pred = ebm.predict(X_test)

# Evaluate the model on the test set
mse_test = mean_squared_error(y_test, y_test_pred)
print(f'Mean Squared Error on Test Set: {mse_test}')
```

```python
#  size of each dataset
print(train_data.shape)
print(test_data.shape)
print(val_data.shape)
```

```python
# Create a DataFrame to store actual and predicted values for the test set
results_df = pd.DataFrame({'Actual': y_test, 'Predicted': y_test_pred})

# Scatter plot of Actual vs. Predicted values with different colors
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Actual', y='Predicted', data=results_df, color='blue', alpha=0.7, label='Predicted')
sns.scatterplot(x='Actual', y='Actual', data=results_df, color='red', alpha=0.7, label='Actual')
plt.title('Actual vs. Predicted Values')
plt.xlabel('Actual Wine Quality')
plt.ylabel('Predicted Wine Quality')
plt.legend()
plt.show()

```
